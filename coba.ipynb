```python
!pip install pandas torch torch-geometric scikit-learn

import pandas as pd
import torch
from torch_geometric.data import Data
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load CSV hasil ekspor
nodes = pd.read_csv("nodes.csv")
edges = pd.read_csv("edges.csv")

# Encode node id (Neo4j id bisa besar)
node_id_map = {nid: i for i, nid in enumerate(nodes['node_id'])}
edges['source'] = edges['source'].map(node_id_map)
edges['target'] = edges['target'].map(node_id_map)

# Siapkan fitur numerik
features = ['tarif_diklaim', 'tarif_seharusnya', 'lama_rawat']
X = nodes[features].fillna(0)
X = StandardScaler().fit_transform(X)
x = torch.tensor(X, dtype=torch.float)

# Label fraud (0/1)
y = torch.tensor(nodes['is_fraud'].fillna(0).astype(int).values, dtype=torch.long)

# Edge index
edge_index = torch.tensor(edges[['source','target']].values.T, dtype=torch.long)

# Buat data graph
data = Data(x=x, edge_index=edge_index, y=y)

print(data)
```
